### Question 1:
Step 1: Express the expected value \( E[y] \) in terms of \( P(y) \):

$E[y] = \sum_{y} y \cdot P(y) $

Step 2: Substitute the given joint probability distribution for \( P(y) \):

$E[y] = \sum_{y} y \cdot \left(\frac{n!}{y_1!y_2!y_3!...y_k!} \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k}\right) $

Step 3: Expand the sum:

$ E[y] = \sum_{y} \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (y_1 \cdot p_1) \cdot (y_2 \cdot p_2) \cdot (y_3 \cdot p_3) \ldots (y_k \cdot p_k) $

Step 4: Rearrange the terms:

$ E[y] =  \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot \sum_{y} (y_1 \cdot p_1) \cdot (y_2 \cdot p_2) \cdot (y_3 \cdot p_3) \ldots (y_k \cdot p_k) $

Step 5: Observe that the sum is a product of sums:

$E[y] =  \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot  \left( \sum_{y_1} \,y_1 \cdot p_1 \right) \cdot \left( \sum_{y_2} \, y_2 \cdot p_2 \right) \cdot \left( \sum_{y_3} \, y_3 \cdot p_3 \right) \ldots \left( \sum_{y_k} \, y_k \cdot p_k \right) $

Step 6: Each individual sum is simply the expected value of a binomial distribution with parameters \( n \) and \( p_i \):

$ E[Y_i] = n \cdot p_i $

Step 7: Substitute the expected values of each \( Y_i \):

$E[y] = \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (n \cdot p_1) \cdot (n \cdot p_2) \cdot (n \cdot p_3) \ldots  (n \cdot p_k)$

Step 8: Notice that \( n! \) and \( y_1!y_2!y_3!...y_k! \) cancel out:

$E[y] = n \cdot p_1 \cdot n \cdot p_2 \cdot n \cdot p_3 \ldots n \cdot p_k $

We simplify the expression by combining terms and recognizing that the coefficient $(y_1!y_2!y_3!...y_k!) $ can be simplified to$ n! $, the number of permutations of outcomes.

Step 9: Simplify:

$E[y] = n  (p_1 \cdot  p_2 \cdot  p_3 \ldots  p_k) $

Step 10: Recall that \( p_1 + p_2 + p_3 + \ldots + p_k = 1 \):

$ E[y] = n \cdot 1 $

$ E[y] = n \cdot p_i $

To conclude, the expected value $E[y]$ equals $\Sigma y \cdot P(y) $, which is equal to $n \cdot p_i$.

$\textbf{Answers for task 2}$

### Question 1
The probability distribution of the experiment can be derived using the binomial distribution formula: $P(Y = y) = \binom{n}{y} \cdot p^y \cdot q^{n-y}$
- $P(Y = y)$ is the probability of observing exactly $y$ successes,
- $n$ is the number of trials,
- $p$ is the probability of success on a single trial,
- $q = (1 - p)$ is the probability of failure on a single trial.


### Question 2

Step 1: Define the probability mass function (PMF) of the binomial distribution:

$ P(Y = y) = \binom{n}{y} \cdot p^y \cdot q^{n-y} $

Where:
- $ n $ is the number of trials,
- $ y $ is the number of successes,
- $ p $ is the probability of success on a single trial,
- $ q = (1 - p) $ is the probability of failure on a single trial.

Step 2: Express the expectation $ E[Y] $ as a sum of $ y \cdot P(Y = y) $ over all possible values of $ y $:

$E[Y] = \sum_{y=0}^{n} y \cdot P(Y = y) $

Step 3: Substitute the expression for $P(Y = y)$ into the expectation formula:

$ E[Y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} \cdot p^y \cdot q^{n-y} $

Step 4: Expand the binomial coefficient $ \binom{n}{y}$ using its definition:

$ E[Y] = \sum_{y=0}^{n} y \cdot \frac{n!}{y!(n-y)!} \cdot p^y \cdot q^{n-y}$

Step 5: Rearrange terms:

$E[Y] = \sum_{y=0}^{n} \frac{n!}{(y-1)!(n-y)!} \cdot p^y \cdot q^{n-y} $

Step 6: Rewrite the sum starting from $ y = 1 $, since the term $ y = 0 $ contributes zero to the sum:

$ E[Y] = \sum_{y=1}^{n} \frac{n!}{(y-1)!(n-y)!} \cdot p^y \cdot q^{n-y} $

Step 7: Factor out $ np $ from the sum:

$E[Y] = np \cdot \sum_{y=1}^{n} \frac{(n-1)!}{(y-1)!(n-1-(y-1))!} \cdot p^{y-1} \cdot q^{n-1-(y-1)} $

Step 8: Rewrite the sum using the binomial coefficient $\binom{n-1}{y-1} $:

$E[Y] = np \cdot \sum_{y=0}^{n-1} \frac{(n-1)!}{y!(n-1-y)!} \cdot p^{y} \cdot q^{n-1-y}$

Step 9: Note that $\sum_{y=0}^{n-1} \binom{n-1}{y} \cdot p^{y} \cdot q^{n-1-y} $ is simply $(p + q)^{n-1} $, which equals 1. 

Step 10: Simplify:

$ E[Y] = np \cdot 1^{n-1} = np $

So, we've shown that the expectation $ E[Y] $ of this probability distribution is indeed $E[Y] = \Sigma y \cdot P(y) = np $.



### Question 3:

$\textbf{Given}:$
- The probability of a person recovering from the illness without the medication is $ p = 0.3 $.
- The number of trials (people) is $n = 10 $.
- The number of people who recovered is $ k = 9 $.

$text\bf{Assumption}:$
- The medication is worthless, implying that the probability of recovery remains $ p = 0.3 $.

$text\bf{Problem:}$
Calculate the probability that at least 9 out of 10 people recover by chance alone.

$text\bf{Approach:}$
We use the binomial distribution. The binomial probability $ P(X = k)$ for $ k $ successes in $ n $ trials is given by:
\begin{gather}
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} 
\end{gather}
First, compute the probability of exactly 9 people recovering:
\begin{gather}
P(X = 9) = \binom{10}{9} (0.3)^9 (0.7)^1 
\end{gather}

Next, compute the probability of exactly 10 people recovering:
\begin{gather}
P(X = 10) = \binom{10}{10} (0.3)^{10} (0.7)^0 
\end{gather}

Finally, sum these probabilities to get the probability of at least 9 people recovering:
\begin{gather}
P(X \geq 9) = P(X = 9) + P(X = 10) 
\end{gather}

### Calculation:
1. - $ P(X = 9) = \binom{10}{9} (0.3)^9 (0.7)^1 $
   - $ \binom{10}{9} = 10 $
   - $ (0.3)^9 = 1.9683 \times 10^{-5} $
   - $ (0.7)^1 = 0.7 $
   - $ P(X = 9) = 10 \times 1.9683 \times 10^{-5} \times 0.7 $
   - $ P(X = 9) = 10 \times 1.37781 \times 10^{-5}$
   - $ P(X = 9) = 1.37781 \times 10^{-4} $

2. - $ P(X = 10) = \binom{10}{10} (0.3)^{10} (0.7)^0 $
   - $\binom{10}{10} = 1 $
   - $ (0.3)^{10} = 5.9049 \times 10^{-6} \$
   - $ (0.7)^0 = 1 $
   - $P(X = 10) = 1 \times 5.9049 \times 10^{-6} $
   - $P(X = 10) = 5.9049 \times 10^{-6} $

3. - $ P(X \geq 9) = P(X = 9) + P(X = 10) $
   - $P(X \geq 9) = 1.37781 \times 10^{-4} + 5.9049 \times 10^{-6} $
   - $ P(X \geq 9) = 1.43686 \times 10^{-4} $

### Conclusion:
The probability that at least 9 out of 10 people will recover by chance alone, assuming the medication is worthless, is:

\begin{gather} 
P(X \geq 9) = 1.43686 \times 10^{-4} \approx 0.000144 
\end{gather}

So, the probability is approximately $ 0.0144\% $.

### Question 1:


The given probability distribution is the Poisson distribution:

\begin{gather}
p(y) = \frac{\lambda^y}{y!} e^{-\lambda}
\end{gather}

To find the expected value $ E[Y] $ of this distribution, we use the definition of the expected value for a discrete random variable:

\begin{gather}
E[Y] = \sum_{y=0}^{\infty} y \cdot p(y)
\end{gather}

Substituting the Poisson probability mass function $ p(y) $:

\begin{gather}
E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}
\end{gather}

Let's simplify this expression. Notice that $ y $ can be factored out:

\begin{gather}
E[Y] = e^{-\lambda} \sum_{y=0}^{\infty} \frac{y \lambda^y}{y!}
\end{gather}

We can rewrite the term involving $ y $ by considering $ y = 0 $ separately (since $ y $ starts at 0 but contributes nothing to the sum):

\begin{gather}
E[Y] = e^{-\lambda} \sum_{y=1}^{\infty} \frac{y \lambda^y}{y!}
\end{gather}

Next, simplify $ \frac{y \lambda^y}{y!} $:

\begin{gather}
\frac{y \lambda^y}{y!} = \frac{\lambda^y}{(y-1)! \cdot y} = \lambda \cdot \frac{\lambda^{y-1}}{(y-1)!}
\end{gather}

Thus, we have:

\begin{gather}
E[Y] = \lambda e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!}
\end{gather}

Now, let $ k = y-1 $. When $ y = 1 $, $ k = 0 $, and when $ y \to \infty $, $ k \to \infty $. Therefore, the summation index change gives us:

\begin{gather}
E[Y] = \lambda e^{-\lambda} \sum_{k=0}^{\infty} \frac{\lambda^k}{k!}
\end{gather}

Recognize that the sum $ \sum_{k=0}^{\infty} \frac{\lambda^k}{k!} $ is the Taylor series expansion for $ e^{\lambda} $:

\begin{gather}
E[Y] = \lambda e^{-\lambda} e^{\lambda} \\
E[Y] = \lambda
\end{gather}

Thus, the expected value $ E[Y] $ for the Poisson distribution is:

\begin{gather}
E[Y] = \lambda
\end{gather}
